#  Analysis of the effect of age on the location of cancer

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv ("E:\Momeni-Machne Learning class momeni 21 01 1402\data science/metadata.csv")
data
```

```python
df = pd.DataFrame (data)
df
```

```python
df.describe()
```

```python
plt.scatter (df['age'], df['localization'])
```

```python
df2 = df.dropna ()
df2
```

```python
plt.scatter (df2['age'],df2['localization'])
plt.xlabel('Age')
plt.ylabel('cancer Localization')
plt.grid()
plt.show()
```

```python
df3 = df2.sort_values(by=['age'])
```

```python
plt.scatter (df3['age'],df3['localization'])
plt.xlabel('Age')
plt.ylabel('cancer Localization')
plt.grid()
plt.show()
```

```python
plt.scatter (df3['sex'],df3['localization'])
plt.xlabel('Gender')
plt.ylabel('Cancer Localization')
plt.grid()
plt.show()
```

```python
plt.bar (df3['age'],df3['localization'])
plt.xlabel('Age')
plt.ylabel('cancer Localization')
plt.show()
```
